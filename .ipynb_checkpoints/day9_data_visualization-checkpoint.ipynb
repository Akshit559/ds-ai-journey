```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Enable inline display in Jupyter
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='blue', marker='o', linestyle='--')
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
students = ["Akshit", "Riya", "Aman", "Pragati"]
marks = [88, 92, 76, 85]

plt.bar(students, marks, color='skyblue')
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
activities = ["Study", "Workout", "Sleep", "Leisure"]
time_spent = [6, 2, 8, 4]

plt.pie(time_spent, labels=activities, autopct="%1.1f%%", startangle=90)
plt.title("Daily Routine Breakdown")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title("Distribution of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample dataset
df = sns.load_dataset("tips")
print(df.head())

# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
sns.histplot(df["total_bill"], bins=20, kde=True, color="teal")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
sns.countplot(x="day", data=df, palette="viridis")
plt.title("Count of Customers by Day")
plt.show()
```

```python
sns.pairplot(df, hue="sex")
plt.show()
```
